```python
import re
import pandas as pd
```

```python
f=open("WhatsApp Chat with Salma Sis.txt",'r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
pattern="\d{1,2}\/\d{2,4}\/\d{2,4},\s\d{1,2}:\d{1,2}\s\w{1,2}\s-\s"
```

```python
messages=re.split(pattern,data)[1:]
messages
```

```python
date2=re.findall(pattern,data)
dates=[]
for i in date2:
    dates.append(i.encode('ascii', 'ignore').decode("utf-8"))
dates
```

```python
df=pd.DataFrame({'user_message':messages,'message_date':dates})
#convert message_Data type
df['message_data']=pd.to_datetime(df['message_date'],format='%m/%d/%y, %H:%M%p - ')
df.rename(columns={'message_date':'date'},inplace=True)
df
```
